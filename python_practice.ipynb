```python
# 請撰寫一程式，讓使用者輸入一個正數s，代表正五邊形之邊長，計算並輸出此正五邊形之面積（Area）。

'''
提示1：建議使用import math模組的math.pow及math.tan
提示2：正五邊形面積的公式： $ Area=(5 * s^2)/(4 * tan(pi/5))$
提示3：輸出浮點數到小數點後第四位。

'''


import math

def pentagon(a):
    area = (1/4)*math.sqrt(a*(a + 2*math.sqrt(a)))*a**2
    return area

a = int(input(a))
area = pentagon(a)

print(f'Area = {area:.4f}')
```

```python
# 正n邊形面積計算
'''
設計說明：
請撰寫一程式，讓使用者輸入兩個正數n、s，代表正n邊形之邊長為s，計算並輸出此正n邊形之面積（Area）。

提示1：建議使用import math模組的math.pow及math.tan
提示2：正n邊形面積的公式如下：
提示3：輸出浮點數到小數點後第四位

'''

def polgon(n,s):
    area = (n*s**2)/(4*math.tan(math.pi/n))
    return area 

n = int(input())
s = int(input())

area = polgon(n,s)

print(f'Area = {area:.4f}')
```

```python
# 偶數判斷
'''
設計說明：請使用選擇敘述撰寫一程式，讓使用者輸入一個正整數，然後判斷它是否為偶數（even）。
'''

num = int(input())

if num % 2 == 0 :
    print(f'{num} is an even number')
else:
     print(f'{num} is not  an even number')
```

```python
# 倍數判斷
'''
設計說明：
請使用選擇敘述撰寫一程式，讓使用者輸入一個正整數，然後判斷它是3或5的倍數，顯示【x is a multiple of 3.】或【x is a multiple of 5.】；若此數值同時為3與5的倍數，顯示【x is a multiple of 3 and 5.】；如此數值皆不屬於3或5的倍數，顯示【x is not a multiple of 3 or 5.】，將使用者輸入的數值代入x。

'''

num = int(input())

if num%3 == 0 and num%5 != 0:
    print(f'{num} is a multiple of 3.')
elif num%3 != 0 and num%5 == 0 :
    print(f'{num} is a multiple of 5.')
elif num%3 == 0 and num%5 == 0 :
    print(f'{num} is a multiple of 3 and 5.')
else:
    print(f'{num} is not a multiple of 3 or 5.')
```

```python
# 閏年判斷
'''
設計說明：
請使用選擇敘述撰寫一程式，讓使用者輸入一個西元年份，然後判斷它是否為閏年（leap year）或平年。其判斷規則為：每四年一閏，每百年不閏，但每四百年也一閏。

'''

year = int(input())

if year%4 == 0 and ( year%400 ==0 or year != 100 ):
    print(f'{year} is a leap year.')
else : 
    print(f'{year} is not a leap year.')
```

```python
# 算術運算
'''
設計說明：
請使用選擇敘述撰寫一程式，讓使用者輸入兩個整數a、b，然後再輸入一算術運算子 (+、-、*、/、//、%） ，輸出經過運算後的結果。
'''

num1 = int(input())
num2 = int(input())
opeation = input()

if opeation == '+':
    print(num1*num2)
elif opeation == '-':
    print(num1 - num2)
elif opeation == '*':
    print(num1*num2)
elif opeation == '/':
    print(num1/num2)
elif opeation == '//':
    print(num1//num2)
elif opeation == '%':
    print(num1%num2)
else:
    print('請輸入算術運算子')
```

```python
# 等級判斷
'''
設計說明：
請使用選擇敘述撰寫一程式，根據使用者輸入的分數顯示對應的等級。標準如下表所示：
分　數 等級
80 ~ 100 A
70 ~ 79 B
60 ~ 69 C
<= 59 F
'''

score = int(input())

if score >= 80 and  score <= 100:
    print('A')
elif score >= 70 and  score <= 79:
    print('B')
elif score >= 60 and  score <= 69:
    print('C')
else : 
    print('F')
```

```python
# 折扣方案
'''
設計說明：
請使用選擇敘述撰寫一程式，要求使用者輸入購物金額，購物金額需大於8,000（含）以上，並顯示折扣優惠後的實付金額。購物金額折扣方案如下表所示：
金額 折扣
8,000(含)以上 9.5折
18,000（含）以上 9折
28,000（含）以上 8折
38,000（含）以上 7折
'''
money = int(input())

if money < 8000:
    print('購買金額需要大於8,000')
elif money >= 8000 and money < 18000:
    print(money*0.95)
elif money >= 18000 and money < 28000:
    print(money*0.9)
elif money >= 28000 and money < 38000:
    print(money*0.8)
elif money >= 38000 :
    print(money*0.7)
```

```python
# 距離判斷
'''
設計說明：
請使用選擇敘述撰寫一程式，讓使用者輸入一個點的平面座標x和y值，判斷此點是否與點(5, 6)的距離小於或等於15，如距離小於或等於15顯示【Inside】，反之顯示【Outside】。

'''

import math

x2 = float(input())
y2 = float(input())

if math.sqrt((x2-5)**2 + (y2-6)**2) <= 15 :
    print('Inside')
else : 
    print('Outside')
```

```python
# 三角形判斷
'''
設計說明：
請使用選擇敘述撰寫一程式，讓使用者輸入三個邊長，檢查這三個邊長是否可以組成一個三角形。若可以，則輸出該三角形之周長；否則顯示【Invalid】。
提示：檢查方法 = 任意兩個邊長之總和大於第三邊長
'''

a = int(input())
b = int(input())
c = int(input())

if a+b > c and a+c > b and b+c > a :
    print(a+b+c)
else:
    print('Invalid')
```
