```python
import pandas as pd 

pd.set_option('display.max_columns',None)

df = pd.read_csv('./archive/qs-world-university-rankings-2017-to-2022-V2.csv')

df[df['international_students'] == '3.705'] 
```

```python
df[['university','year']]

df[df['year'] == 2017]

df['international_students'] = df['international_students'].fillna(0).astype(str)
# df['international_students'].astype(str)
# print(df['international_students'].dtype)
df.isna().sum()
```

```python
print(df['international_students'].dtype)
print(df['international_students'].isna().sum())

df['international_students'] = df['international_students'].str.replace(',','')
df['international_students'] = df['international_students'].str.replace('.','')

print(df['international_students'].isna().sum())

df['international_students'] = df['international_students'].astype(int)

print(df['international_students'].dtype)


```

```python
print(df['type'].unique())

def engTochi(s):
    if s == 'Private':
        s = '私立'
    elif s == 'Public':
        s = '公立'
    else:
        s = '未知'
    return s 

df['type'] = df['type'].apply(engTochi)

# df['type'] = df['type'].map({'Private':'私立','Public':'公立'})
# df['type'] = df['type'].fillna('未知')
# df['type'].unique()

df[['university','type']]
```

```python
df['logo_pic_format'] = df['logo'].apply(lambda x : x.split('.')[-1])
df   
```

```python
df['international_students_level'] = pd.cut(df['international_students'],bins=3,labels=['少','中','多'])

df[['university','international_students','international_students_level']]

df[df['international_students_level'] == '多'][['university','international_students','international_students_level']].head(10)

```
